##Projeto 1 - Algoritmo de Ordenação Bolha (Bubble Sort)

Abaixo estão os algorítmos para ordenação, utilizando o método bolha de ordenação, com entradas de vetores de tamanhos 1.000, 10.000 e 100.000
- aleatórias
- ordenadas em forma crescente
- ordenadas em forma decrescente

Abaixo de cada algorítmo estão os tempos observados somando 4 tentativas.

##Vetor em ordem aleatória com 1000 elementos:



```python
import requests
import time

# Importar dados
url = "https://raw.githubusercontent.com/anabiarochar/Mestrado_AlgoritmosProgramacao/main/Projeto1/Vetores/vetor_1000_aleatorio.txt"

try:
    response = requests.get(url)
    content = response.text

    # Ler arquivo como vetor separado por vírgula
    alist = [int(x.strip()) for x in content.split(',') if x.strip()]

except Exception as e:
    print(f"Erro ao carregar ou processar o arquivo: {e}")

# Método Bolha
def bubbleSort(alist):
  for passnum in range(len(alist)-1,0,-1):
    for i in range(passnum):
      if alist[i]>alist[i+1]:
        temp = alist[i]
        alist[i] = alist[i+1]
        alist[i+1] = temp


# Execução e Tempo
if alist:
    inicio = time.perf_counter()
    vetor_ordenado = bubbleSort(alist)
    fim = time.perf_counter()

    tempo_total = fim - inicio

    print(f"Tempo de execução: {tempo_total:.6f} segundos")
    print(alist[:10])

else:
    print("Vetor vazio")
```

###Tempo de Execução:

Tentativa 1 - Tempo de execução: 0.083225 segundos   
Tentativa 2 - Tempo de execução: 0.051019 segundos  
Tentativa 3 - Tempo de execução: 0.050137 segundos  
Tentativa 4 - Tempo de execução: 0.050899 segundos
