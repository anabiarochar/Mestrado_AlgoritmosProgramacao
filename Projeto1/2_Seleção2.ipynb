##Projeto 1 - Algoritmo de Ordenação Seleção (Selection Sort)

Abaixo estão os algorítmos para ordenação, utilizando o método seleção, de entradas de vetores de tamanhos 1.000, 10.000 e 100.000
- aleatórias
- ordenadas em forma crescente
- ordenadas em forma decrescente

Abaixo de cada algorítmo estão os tempos observados somando 4 tentativas.

```python
import requests
import time

# Importar dados
url = "https://raw.githubusercontent.com/anabiarochar/Mestrado_AlgoritmosProgramacao/main/Projeto1/Vetores/vetor_1000_aleatorio.txt"

try:
    response = requests.get(url)
    content = response.text

    # Ler arquivo como vetor separado por vírgula
    vetor = [int(x.strip()) for x in content.split(',') if x.strip()]

except Exception as e:
    print(f"Erro ao carregar ou processar o arquivo: {e}")


# Método de Seleção
def selectionSort(vetor):
  n = len(vetor)
  for i in range(n):
      min = i
      for j in range(i + 1, n):
            if vetor[j] < vetor[min]:
                min = j

      vetor[i], vetor[min] = vetor[min], vetor[i]

  return vetor

# Execução e Tempo
if vetor:
    inicio = time.perf_counter()
    vetor_ordenado = selectionSort(vetor)
    fim = time.perf_counter()

    tempo_total = fim - inicio

    print(f"Tempo de execução: {tempo_total:.6f} segundos")
    print(vetor[:10])

else:
    print("Vetor vazio")

```
