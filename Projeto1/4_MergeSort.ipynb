```python
import requests
import time

# Importar dados
url = "https://raw.githubusercontent.com/anabiarochar/Mestrado_AlgoritmosProgramacao/refs/heads/main/Projeto1/Vetores/vetor_100000_aleatorio.txt?token=GHSAT0AAAAAADXG3NI4RBTUJ2LPBXXEFEEC2NOIGFQ"

vetor = []

try:
    response = requests.get(url)
    content = response.text

    # Ler arquivo como vetor separado por vírgula
    vetor = [int(x.strip()) for x in content.split(',') if x.strip()]

except Exception as e:
    print(f"Erro ao carregar o arquivo: {e}")

# Método Merge Sort
def mergeSort(arr):
    n = len(arr) # n=len(V)
    if n <= 1:
        return arr

    meio = n // 2

    esquerda = []
    for i in range(meio):
        esquerda.append(arr[i])

    direita = []
    for i in range(meio, n):
        direita.append(arr[i])

    esquerda = mergeSort(esquerda)
    direita = mergeSort(direita)

    return merge(esquerda, direita)

def merge(esquerda, direita):
    resultado = []
    i = j = 0

    while i < len(esquerda) and j < len(direita):
        if esquerda[i] < direita[j]:
            resultado.append(esquerda[i])
            i += 1
        else:
            resultado.append(direita[j])
            j += 1

    resultado.extend(esquerda[i:])
    resultado.extend(direita[j:])
    return resultado

# Execução e Tempo
if vetor:

    inicio = time.perf_counter()
    vetor_ordenado = mergeSort(vetor)
    fim = time.perf_counter()

    tempo_total = fim - inicio

    print(f"Tempo de execução: {tempo_total:.6f} segundos")
    print(vetor_ordenado[:10])
else:
    print("Não foi possível processar o vetor.")
```
