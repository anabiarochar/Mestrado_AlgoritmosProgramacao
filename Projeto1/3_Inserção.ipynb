```python
import requests
import time

# Importar dados
#url = "https://raw.githubusercontent.com/anabiarochar/Mestrado_AlgoritmosProgramacao/main/Projeto1/Vetores/vetor_100000_decrescente.txt"
url = "https://raw.githubusercontent.com/anabiarochar/Mestrado_AlgoritmosProgramacao/refs/heads/main/Projeto1/Vetores/vetor_1000_aleatorio.txt?token=GHSAT0AAAAAADXG3NI425JILOEHF3L3VXEI2NOH27Q"

try:
    response = requests.get(url)
    content = response.text

    # Ler arquivo como vetor separado por vírgula
    vetor = [int(x.strip()) for x in content.split(',') if x.strip()]

except Exception as e:
    print(f"Erro ao carregar ou processar o arquivo: {e}")

# Método Inserção
def insertionSort(A, n):
    for j in range(1, n):
        x = A[j]
        i = j - 1

        while i >= 0 and A[i] > x:
            A[i + 1] = A[i]
            i -= 1
        A[i + 1] = x
    return A

# Execução e Tempo
if vetor:

    inicio = time.perf_counter()
    vetor_para_ordenar = vetor.copy()
    vetor_ordenado = insertionSort(vetor_para_ordenar, len(vetor_para_ordenar))
    fim = time.perf_counter()

    tempo_total = fim - inicio

    print(f"Tempo de execução: {tempo_total:.6f} segundos")
    print(vetor_ordenado[:10])
else:
    print("Não foi possível processar o vetor.")
```
