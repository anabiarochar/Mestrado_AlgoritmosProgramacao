##Projeto 1 - Algoritmo de Ordenação Inserção (Insertion Sort)

Abaixo estão os algorítmos para ordenação, utilizando o método inserção de ordenação, com entradas de vetores de tamanhos 1.000, 10.000 e 100.000
- aleatórias
- ordenadas em forma crescente
- ordenadas em forma decrescente

Abaixo de cada algorítmo estão os tempos observados somando 4 tentativas.

##Vetor em ordem aleatória com 1000 elementos:


```python
import requests
import time

# Importar dados
url = "https://raw.githubusercontent.com/anabiarochar/Mestrado_AlgoritmosProgramacao/main/Projeto1/Vetores/vetor_1000_aleatorio.txt"

vetor = []

try:
    response = requests.get(url)
    content = response.text

    # Ler arquivo como vetor separado por vírgula
    vetor = [int(x.strip()) for x in content.split(',') if x.strip()]

except Exception as e:
    print(f"Erro ao carregar ou processar o arquivo: {e}")

# Método Inserção
def insertionSort(A, n):
    for j in range(1, n):
        x = A[j]
        i = j - 1

        while i >= 0 and A[i] > x:
            A[i + 1] = A[i]
            i -= 1
        A[i + 1] = x
    return A

# Execução e Tempo
if vetor:

    inicio = time.perf_counter()
    vetor_para_ordenar = vetor.copy()
    vetor_ordenado = insertionSort(vetor_para_ordenar, len(vetor_para_ordenar))
    fim = time.perf_counter()

    tempo_total = fim - inicio

    print(f"Tempo de execução: {tempo_total:.6f} segundos")
    print(vetor_ordenado[:10])
else:
    print("Não foi possível processar o vetor.")
```

###Tempo de Execução:

Tentativa 1 - Tempo de execução: 0.025019 segundos  
Tentativa 2 - Tempo de execução: 0.024606 segundos  
Tentativa 3 - Tempo de execução: 0.021900 segundos  
Tentativa 4 - Tempo de execução: 0.048388 segundos
