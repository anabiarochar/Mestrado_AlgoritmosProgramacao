```python
import requests
import time

# 1. Obter os dados do GitHub
url = "https://raw.githubusercontent.com/anabiarochar/Mestrado_AlgoritmosProgramacao/refs/heads/main/Projeto1/Vetores/vetor_10000_aleatorio.txt?token=GHSAT0AAAAAADXG3NI43XWVLIM2PVGKVOFK2NN5GXQ"

try:
    response = requests.get(url)
    content = response.text

    # Lendo o arquivo separado por vírgulas
    vetor = [int(x.strip()) for x in content.split(',') if x.strip()]

except Exception as e:
    print(f"Erro ao carregar ou processar o arquivo: {e}")


# 2. Método de Seleção
def selectionSort(vetor):
  n = len(vetor)
  for i in range(n):
      min = i
      for j in range(i + 1, n):
            if vetor[j] < vetor[min]:
                min = j

      # Trocar o menor elemento encontrado com o elemento da posição i
      vetor[i], vetor[min] = vetor[min], vetor[i]

  return vetor

# 3. Execução e Medição de Tempo
if vetor:
    inicio = time.perf_counter()
    vetor_ordenado = selectionSort(vetor)
    fim = time.perf_counter()

    tempo_total = fim - inicio

    print(f"Tempo de execução: {tempo_total:.6f} segundos")
    print(vetor[:10])

else:
    print("O vetor está vazio. Verifique o link ou o formato do arquivo.")

```
