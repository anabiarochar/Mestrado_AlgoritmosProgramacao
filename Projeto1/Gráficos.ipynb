```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# 1. Preparação dos dados (Exemplo de dados coletados)
# Substitua os valores abaixo pelos tempos reais obtidos nos seus testes
dados = {
    'Tamanho do Vetor': [1000, 10000, 100000] * 5,
    'Tempo (s)': [
        0.002, 0.02, 0.2,   # MergeSort
        0.001, 0.015, 0.15, # QuickSort
        0.05,  5.2,   500,  # BubbleSort (estimado para 100k)
        0.03,  3.1,   310,  # InsertionSort
        0.04,  4.5,   450   # SelectionSort
    ],
    'Algoritmo': (['MergeSort']*3 + ['QuickSort']*3 + ['BubbleSort']*3 +
                  ['InsertionSort']*3 + ['SelectionSort']*3)
}

df = pd.DataFrame(dados)

# 2. Configuração estética do Seaborn
sns.set_theme(style="whitegrid")
plt.figure(figsize=(12, 7))

# 3. Criação do gráfico de linhas com escala logarítmica
# A escala logarítmica é CRUCIAL aqui, pois os métodos O(n log n)
# são ordens de magnitude mais rápidos que os O(n²)
g = sns.lineplot(data=df, x='Tamanho do Vetor', y='Tempo (s)',
                 hue='Algoritmo', style='Algoritmo', markers=True, dashes=False, linewidth=2.5)

g.set_yscale("log") # Escala logarítmica para visualização justa
g.set_xscale("linear")

# 4. Customização de títulos e eixos
plt.title('Comparação de Desempenho: Algoritmos de Ordenação', fontsize=16, fontweight='bold')
plt.xlabel('Tamanho do Vetor (N)', fontsize=12)
plt.ylabel('Tempo de Execução (segundos) - Escala Log', fontsize=12)
plt.xticks([1000, 10000, 100000])

plt.legend(title='Métodos', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

# Salvar o gráfico para o relatório
plt.savefig('comparativo_algoritmos.png', dpi=300)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Configuração de estilo
sns.set_theme(style="whitegrid")
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# --- DADOS PARA O GRÁFICO 1: MÉTODOS EFICIENTES O(n log n) ---
# Substitua pelos seus tempos reais obtidos no Colab
dados_eficientes = {
    'N': [1000, 10000, 100000, 1000, 10000, 100000],
    'Tempo (s)': [0.0002, 0.0025, 0.031,  # Exemplo Merge Sort
                  0.0001, 0.0018, 0.025], # Exemplo Quick Sort
    'Algoritmo': ['Merge Sort']*3 + ['Quick Sort']*3
}
df_ef = pd.DataFrame(dados_eficientes)

sns.lineplot(data=df_ef, x='N', y='Tempo (s)', hue='Algoritmo', marker='o', ax=ax1, linewidth=2)
ax1.set_title('Algoritmos Eficientes: O(n log n)', fontsize=14, fontweight='bold')
ax1.set_xlabel('Tamanho do Vetor (N)')
ax1.set_ylabel('Tempo (segundos)')

# --- DADOS PARA O GRÁFICO 2: MÉTODOS QUADRÁTICOS O(n²) ---
# Nota: Para 100k, esses tempos podem ser MUITO altos (ex: > 3000s)
dados_quadraticos = {
    'N': [1000, 10000, 100000, 1000, 10000, 100000, 1000, 10000, 100000],
    'Tempo (s)': [0.04, 4.2, 420,   # Exemplo Bubble Sort
                  0.02, 2.1, 210,   # Exemplo Insertion Sort
                  0.03, 3.5, 350],  # Exemplo Selection Sort
    'Algoritmo': ['Bubble Sort']*3 + ['Insertion Sort']*3 + ['Selection Sort']*3
}
df_quad = pd.DataFrame(dados_quadraticos)

sns.lineplot(data=df_quad, x='N', y='Tempo (s)', hue='Algoritmo', marker='s', ax=ax2, palette='viridis', linewidth=2)
ax2.set_title('Algoritmos Quadráticos: O(n²)', fontsize=14, fontweight='bold')
ax2.set_xlabel('Tamanho do Vetor (N)')
ax2.set_ylabel('Tempo (segundos)')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# 1. Dados de exemplo (Substitua pelos seus valores reais medidos)
# Os valores abaixo seguem a proporção esperada da complexidade assintótica
data = {
    'Algoritmo': ['Bubble', 'Bubble', 'Bubble', 'Selection', 'Selection', 'Selection',
                  'Insertion', 'Insertion', 'Insertion', 'Merge', 'Merge', 'Merge',
                  'Quick', 'Quick', 'Quick'],
    'N': [1000, 10000, 100000] * 5,
    'Tempo (s)': [
        0.05, 5.0, 500.0,   # Bubble (O(n²))
        0.03, 3.2, 320.0,   # Selection (O(n²))
        0.02, 2.1, 210.0,   # Insertion (O(n²))
        0.001, 0.012, 0.15, # Merge (O(n log n))
        0.0008, 0.009, 0.11 # Quick (O(n log n))
    ]
}

df = pd.DataFrame(data)

# 2. Configuração da área de desenho (5 gráficos em linha ou 2x3)
fig, axes = plt.subplots(1, 5, figsize=(22, 5), sharex=True)
fig.suptitle('Escalabilidade Individual: Tempo vs Tamanho do Vetor (N)', fontsize=18, fontweight='bold', y=1.05)

algoritmos = ['Bubble', 'Selection', 'Insertion', 'Merge', 'Quick']
cores = ['#e74c3c', '#e67e22', '#f1c40f', '#2ecc71', '#3498db']

# 3. Loop para criar cada subgráfico
for i, algo in enumerate(algoritmos):
    df_filtrado = df[df['Algoritmo'] == algo]
    sns.lineplot(data=df_filtrado, x='N', y='Tempo (s)', ax=axes[i], marker='o', color=cores[i], linewidth=2.5)

    axes[i].set_title(f'Método: {algo}', fontsize=14)
    axes[i].set_xlabel('Tamanho (N)')
    axes[i].set_ylabel('Tempo (segundos)')
    axes[i].grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
