```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# 1. Preparação dos dados
aleatorios = {
    'Tamanho do Vetor': [1000, 10000, 100000] * 5,
    'Tempo (s)': [
        0.003, 0.032, 0.512,       # MergeSort
        0.003, 0.034, 0.449,       # QuickSort
        0.059,  6.877,   763.785,  # BubbleSort (estimado para 100k)
        0.030,  2.849,   428.039,  # InsertionSort
        0.037,  2.724,   350.004   # SelectionSort
    ],
    'Algoritmo': (['MergeSort']*3 + ['QuickSort']*3 + ['BubbleSort']*3 +
                  ['InsertionSort']*3 + ['SelectionSort']*3)
}

df = pd.DataFrame(aleatorios)

# 2. Configuração estética do Seaborn
sns.set_theme(style="whitegrid")
plt.figure(figsize=(12, 7))

# 3. Criação do gráfico de linhas com escala logarítmica
# A escala logarítmica é CRUCIAL aqui, pois os métodos O(n log n)
# são ordens de magnitude mais rápidos que os O(n²)
g = sns.lineplot(data=df, x='Tamanho do Vetor', y='Tempo (s)',
                 hue='Algoritmo', style='Algoritmo', markers=True, dashes=False, linewidth=2.5)

g.set_yscale("log") # Escala logarítmica para visualização justa
g.set_xscale("linear")

# 4. Customização de títulos e eixos
plt.title('Comparação de Desempenho: Algoritmos de Ordenação - Dados Aleatórios', fontsize=16, fontweight='bold')
plt.xlabel('Tamanho do Vetor (N)', fontsize=12)
plt.ylabel('Tempo de Execução (segundos) - Escala Log', fontsize=12)
plt.xticks([1000, 10000, 100000])

plt.legend(title='Métodos', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

# Salvar o gráfico para o relatório
plt.savefig('comparativo_algoritmos_aleatorio.png', dpi=300)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# 1. Dados de exemplo (Substitua pelos seus valores reais medidos)
# Os valores abaixo seguem a proporção esperada da complexidade assintótica
data = {
    'Algoritmo': ['Bubble', 'Bubble', 'Bubble', 'Selection', 'Selection', 'Selection',
                  'Insertion', 'Insertion', 'Insertion', 'Merge', 'Merge', 'Merge',
                  'Quick', 'Quick', 'Quick'],
    'N': [1000, 10000, 100000] * 5,
    'Tempo (s)': [
        0.059, 6.877, 763.785,   # Bubble (O(n²))
        0.037, 2.724, 350.004,   # Selection (O(n²))
        0.030, 2.849, 428.039,   # Insertion (O(n²))
        0.003, 0.034, 0.449,     # Merge (O(n log n))
        0.003, 0.032, 0.512      # Quick (O(n log n))
    ]
}

df = pd.DataFrame(data)

# 2. Configuração da área de desenho (5 gráficos em linha ou 2x3)
fig, axes = plt.subplots(1, 5, figsize=(22, 5), sharex=True)
fig.suptitle('Escalabilidade Individual: Tempo vs Tamanho do Vetor (N)', fontsize=18, fontweight='bold', y=1.05)

algoritmos = ['Bubble', 'Selection', 'Insertion', 'Merge', 'Quick']
cores = ['#e74c3c', '#e67e22', '#f1c40f', '#2ecc71', '#3498db']

# 3. Loop para criar cada subgráfico
for i, algo in enumerate(algoritmos):
    df_filtrado = df[df['Algoritmo'] == algo]
    sns.lineplot(data=df_filtrado, x='N', y='Tempo (s)', ax=axes[i], marker='o', color=cores[i], linewidth=2.5)

    axes[i].set_title(f'Método: {algo}', fontsize=14)
    axes[i].set_xlabel('Tamanho (N)')
    axes[i].set_ylabel('Tempo (segundos)')
    axes[i].grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Dados extraídos da sua tabela
data = {
    'Metodo': ['Bolha', 'Bolha', 'Bolha', 'Seleção', 'Seleção', 'Seleção',
               'Inserção', 'Inserção', 'Inserção', 'Quick Sort', 'Quick Sort', 'Quick Sort',
               'Merge Sort', 'Merge Sort', 'Merge Sort'],
    'Caso': ['Aleatório', 'Crescente', 'Decrescente'] * 5,
    'Tamanho': [100000] * 15, # Focando no cenário mais crítico de 100k
    'Tempo': [763.785, 233.165, 859.368,  # Bolha
              350.004, 287.933, 296.649,  # Seleção
              428.039, 0.091, 609.920,    # Inserção
              0.449, 0.283, 0.271,        # Quick Sort
              0.512, 0.272, 0.272]        # Merge Sort
}

df_100k = pd.DataFrame(data)

# 2. Configuração do gráfico
plt.figure(figsize=(14, 7))
sns.set_theme(style="whitegrid")

# Criando o gráfico de barras
plot = sns.barplot(data=df_100k, x='Metodo', y='Tempo', hue='Caso', palette='viridis')

# 3. APLICANDO ESCALA LOGARÍTMICA
plot.set_yscale("log")

# Customização de títulos e labels
plt.title('Comparação de Performance (N = 100.000) - Escala Logarítmica', fontsize=16, fontweight='bold')
plt.ylabel('Tempo de Execução (segundos) - [LOG]', fontsize=12)
plt.xlabel('Método de Ordenação', fontsize=12)

# Adicionando os valores em cima das barras para facilitar a leitura
for p in plot.patches:
    if p.get_height() > 0:
        plot.annotate(format(p.get_height(), '.3f'),
                      (p.get_x() + p.get_width() / 2., p.get_height()),
                      ha = 'center', va = 'center',
                      xytext = (0, 9),
                      textcoords = 'offset points',
                      fontsize=9, fontweight='bold')

plt.legend(title='Estado Inicial do Vetor', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
