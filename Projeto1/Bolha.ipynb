```python
import requests
import time

# 1. Obter os dados do GitHub
# Nota: Lembre-se que o token no URL pode expirar.
url = "https://raw.githubusercontent.com/anabiarochar/Mestrado_AlgoritmosProgramacao/vetor_1000_aleatorio.txt"

try:
    response = requests.get(url)
    content = response.text

    # AJUSTE AQUI: Separando por vírgula e convertendo para inteiro
    # Usamos .strip() para remover possíveis espaços em branco ao redor dos números
    vetor = [int(x.strip()) for x in content.split(',') if x.strip()]

    print(f"Sucesso! Vetor carregado com {len(vetor)} elementos.")
    print(f"Primeiros 5 elementos lidos: {vetor[:5]}")

except Exception as e:
    print(f"Erro ao carregar ou processar o arquivo: {e}")
    vetor = []

# 2. Implementação do Método Bolha (Bubble Sort)
def bubble_sort(lista_original):
    # Criamos uma cópia para preservar o vetor original
    arr = lista_original.copy()
    n = len(arr)

    for i in range(n):
        trocou = False
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                # Troca os elementos de posição
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                trocou = True

        # Se não houve troca nesta iteração, o vetor já está ordenado
        if not trocou:
            break
    return arr

# 3. Execução e Medição de Tempo
if vetor:
    print("\nIniciando ordenação por Bubble Sort...")

    # Início do cronômetro
    inicio = time.perf_counter()

    vetor_ordenado = bubble_sort(vetor)

    # Fim do cronômetro
    fim = time.perf_counter()

    tempo_total = fim - inicio

    print("--- Resultado ---")
    print(f"Tempo de execução: {tempo_total:.6f} segundos")
    print(f"Primeiros 10 elementos ordenados: {vetor_ordenado[:10]}")
    print(f"Últimos 10 elementos ordenados: {vetor_ordenado[-10:]}")
else:
    print("O vetor está vazio. Verifique o link ou o formato do arquivo.")
```
