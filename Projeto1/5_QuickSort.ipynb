##Projeto 1 - Algoritmo de Ordenação Quick Sort

Abaixo estão os algorítmos para ordenação, utilizando o método quick sort de ordenação, com entradas de vetores de tamanhos 1.000, 10.000 e 100.000

- aleatórias
- ordenadas em forma crescente
- ordenadas em forma decrescente  

Abaixo de cada algorítmo estão os tempos observados somando 4 tentativas.

##Vetor em ordem aleatória com 1000 elementos:

```python
import requests
import time
import random

# Importar dados
url = "https://raw.githubusercontent.com/anabiarochar/Mestrado_AlgoritmosProgramacao/main/Projeto1/Vetores/vetor_1000_aleatorio.txt"

try:
    response = requests.get(url)
    content = response.text
    # Ler arquivo como vetor separado por vírgula
    vetor = [int(x.strip()) for x in content.split(',') if x.strip()]
except Exception as e:
    print(f"Erro ao carregar o arquivo: {e}")

# Método QuickSort com pivô aleatório
def quick_sort_random(arr):
    if len(arr) <= 1:
        return arr

    indice_pivo = random.randint(0, len(arr) - 1)
    pivo = arr[indice_pivo]

    esquerda = [x for i, x in enumerate(arr) if x < pivo]
    meio = [x for x in arr if x == pivo]
    direita = [x for i, x in enumerate(arr) if x > pivo]

    return quick_sort_random(esquerda) + meio + quick_sort_random(direita)

# Execução e Tempo
if vetor:

    inicio = time.perf_counter()
    vetor_ordenado = quick_sort_random(vetor)
    fim = time.perf_counter()

    tempo_total = fim - inicio

    print(f"Tempo de execução: {tempo_total:.6f} segundos")
    print(vetor_ordenado[:10])
else:
    print("Não foi possível processar o vetor.")
```

###Tempo de Execução:

Tentativa 1 - Tempo de execução: 0.003436 segundos  
Tentativa 2 - Tempo de execução: 0.002381 segundos  
Tentativa 3 - Tempo de execução: 0.002268 segundos  
Tentativa 4 - Tempo de execução: 0.002113 segundos
