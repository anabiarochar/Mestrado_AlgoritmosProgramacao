```python

import requests
import time

# Importar dados
url = "https://raw.githubusercontent.com/anabiarochar/Mestrado_AlgoritmosProgramacao/refs/heads/main/Projeto1/Vetores/vetor_10000_aleatorio.txt?token=GHSAT0AAAAAADXG3NI43XWVLIM2PVGKVOFK2NN5GXQ"

try:
    response = requests.get(url)
    content = response.text

    # Ler arquivo como vetor separado por vírgula
    alist = [int(x.strip()) for x in content.split(',') if x.strip()]

except Exception as e:
    print(f"Erro ao carregar ou processar o arquivo: {e}")

# 2. Método Bolha
def bubbleSort(alist):
  for passnum in range(len(alist)-1,0,-1):
    for i in range(passnum):
      if alist[i]>alist[i+1]:
        temp = alist[i]
        alist[i] = alist[i+1]
        alist[i+1] = temp


# 3. Execução e Medição de Tempo
if alist:
    inicio = time.perf_counter()
    vetor_ordenado = bubbleSort(alist)
    fim = time.perf_counter()

    tempo_total = fim - inicio

    print(f"Tempo de execução: {tempo_total:.6f} segundos")
    print(alist[:10])

else:
    print("O vetor está vazio. Verifique o link ou o formato do arquivo.")
```
