```python
import requests
import time

# 1. Obter os dados do GitHub
url = "https://raw.githubusercontent.com/anabiarochar/Mestrado_AlgoritmosProgramacao/refs/heads/main/Projeto1/Vetores/vetor_1000_decrescente.txt?token=GHSAT0AAAAAADXG3NI5M2O4MPNENAJ74TN62NNF5OQ"

try:
    response = requests.get(url)
    content = response.text

    # Lendo o arquivo separado por vírgulas
    vetor = [int(x.strip()) for x in content.split(',') if x.strip()]

    print(f"Sucesso! Vetor carregado com {len(vetor)} elementos.")

except Exception as e:
    print(f"Erro ao carregar ou processar o arquivo: {e}")
    vetor = []

# 2. Implementação do Método de Seleção (Selection Sort)
def selection_sort(lista_original):
    # Criamos uma cópia para preservar o vetor original
    arr = lista_original.copy()
    n = len(arr)

    for i in range(n):
        # Encontrar o índice do menor elemento no restante do vetor
        indice_min = i
        for j in range(i + 1, n):
            if arr[j] < arr[indice_min]:
                indice_min = j

        # Trocar o menor elemento encontrado com o elemento da posição i
        arr[i], arr[indice_min] = arr[indice_min], arr[i]

    return arr

# 3. Execução e Medição de Tempo
if vetor:
    print("\nIniciando ordenação por Selection Sort...")

    # Início do cronômetro de alta precisão
    inicio = time.perf_counter()

    vetor_ordenado = selection_sort(vetor)

    # Fim do cronômetro
    fim = time.perf_counter()

    tempo_total = fim - inicio

    print("--- Resultado ---")
    print(f"Tempo de execução: {tempo_total:.6f} segundos")
    print(f"Primeiros 10 elementos ordenados: {vetor_ordenado[:10]}")
    print(f"Últimos 10 elementos ordenados: {vetor_ordenado[-10:]}")
else:
    print("Não foi possível processar o vetor.")
```
