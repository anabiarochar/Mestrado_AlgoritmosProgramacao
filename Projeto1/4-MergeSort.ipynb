```python
import requests
import time
#import sys

# Aumentar o limite de recursão para garantir que vetores muito grandes não causem erro
#sys.setrecursionlimit(2000)

# 1. Obter os dados do GitHub
url = "https://raw.githubusercontent.com/anabiarochar/Mestrado_AlgoritmosProgramacao/refs/heads/main/Projeto1/Vetores/vetor_100000_decrescente.txt?token=GHSAT0AAAAAADXG3NI4APHQFHXIWU3DHIG62NN3LTQ"

try:
    response = requests.get(url)
    content = response.text
    vetor = [int(x.strip()) for x in content.split(',') if x.strip()]
    print(f"Sucesso! Vetor carregado com {len(vetor)} elementos.")
except Exception as e:
    print(f"Erro ao carregar o arquivo: {e}")
    vetor = []

# 2. Implementação do Método Merge Sort
def merge_sort(arr):
    if len(arr) <= 1:
        return arr

    # Divide o vetor ao meio
    meio = len(arr) // 2
    esquerda = arr[:meio]
    direita = arr[meio:]

    # Chamada recursiva para cada metade
    esquerda = merge_sort(esquerda)
    direita = merge_sort(direita)

    # Junta (interpola) as metades ordenadas
    return merge(esquerda, direita)

def merge(esquerda, direita):
    resultado = []
    i = j = 0

    # Compara os elementos de cada metade e adiciona o menor ao resultado
    while i < len(esquerda) and j < len(direita):
        if esquerda[i] < direita[j]:
            resultado.append(esquerda[i])
            i += 1
        else:
            resultado.append(direita[j])
            j += 1

    # Adiciona os elementos restantes (se houver)
    resultado.extend(esquerda[i:])
    resultado.extend(direita[j:])
    return resultado

# 3. Execução e Medição de Tempo
if vetor:
    print("\nIniciando ordenação por Merge Sort...")

    inicio = time.perf_counter()
    vetor_ordenado = merge_sort(vetor)
    fim = time.perf_counter()

    tempo_total = fim - inicio

    print("--- Resultado ---")
    print(f"Tempo de execução: {tempo_total:.6f} segundos")
    print(f"Primeiros 10 elementos: {vetor_ordenado[:10]}")
else:
    print("Não foi possível processar o vetor.")
```
