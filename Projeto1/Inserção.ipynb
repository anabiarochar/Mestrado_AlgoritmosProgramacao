```python
import requests
import time

# 1. Obter os dados do GitHub
url = "https://raw.githubusercontent.com/anabiarochar/Mestrado_AlgoritmosProgramacao/refs/heads/main/Projeto1/Vetores/vetor_100000_decrescente.txt?token=GHSAT0AAAAAADXG3NI5MPMPFAIQCL7U5CGI2NNGDRQ"

try:
    response = requests.get(url)
    content = response.text

    # Lendo o arquivo separado por vírgulas
    vetor = [int(x.strip()) for x in content.split(',') if x.strip()]

    print(f"Sucesso! Vetor carregado com {len(vetor)} elementos.")

except Exception as e:
    print(f"Erro ao carregar ou processar o arquivo: {e}")
    vetor = []

# 2. Implementação do Método de Inserção (Insertion Sort)
def insertion_sort(lista_original):
    # Criamos uma cópia para preservar o vetor original
    arr = lista_original.copy()
    n = len(arr)

    # Começamos do segundo elemento (índice 1)
    for i in range(1, n):
        chave = arr[i]
        j = i - 1

        # Move os elementos de arr[0..i-1] que são maiores que a chave
        # para uma posição à frente de sua posição atual
        while j >= 0 and chave < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1

        # Insere a chave na posição correta
        arr[j + 1] = chave

    return arr

# 3. Execução e Medição de Tempo
if vetor:
    print("\nIniciando ordenação por Insertion Sort...")

    # Início do cronômetro de alta precisão
    inicio = time.perf_counter()

    vetor_ordenado = insertion_sort(vetor)

    # Fim do cronômetro
    fim = time.perf_counter()

    tempo_total = fim - inicio

    print("--- Resultado ---")
    print(f"Tempo de execução: {tempo_total:.6f} segundos")
    print(f"Primeiros 10 elementos ordenados: {vetor_ordenado[:10]}")
    print(f"Últimos 10 elementos ordenados: {vetor_ordenado[-10:]}")
else:
    print("Não foi possível processar o vetor.")
```
